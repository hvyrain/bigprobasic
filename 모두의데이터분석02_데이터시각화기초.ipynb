```python
# p70, 매년 생일의 그래프 그리기
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding='cp949')
data = csv.reader(f, delimiter=',')
next(data) # 헤더 한 줄 건너띄기

result = []
for row in data :
  if row[4] != '' :
    if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '10' :
      result.append(float(row[4]))
plt.figure(figsize=(12, 4))
plt.title('highest temperature')
plt.plot(result, 'r')
plt.show()
f.close()
```

```python
# p70, 매년 생일의 최저, 최고, 평균 기온 그래프 그리기
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding='cp949')
data = csv.reader(f, delimiter=',')
next(data) # 헤더 한 줄 건너띄기

high = []
low = []
avg = []
for row in data :
  if row[3] != '' and row[4] != '' :
    if 1968 <= int(row[0].split('-')[0]) :
      if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '10' :
        high.append(float(row[4]))
        low.append(float(row[3]))
        avg.append(float(row[2]))
plt.figure(figsize=(12, 4))
plt.rcParams['axes.unicode_minus'] = False
plt.title('highest temperature')
plt.plot(high, 'r', label='high')
plt.plot(low, 'b', label='low')
plt.plot(avg, 'g', label='mean')
plt.legend()
plt.show()
f.close()
```

## 10년씩 기온 평균 구기
- 10년 단위로 평균 기온을 평균내어 그래프로 그려본다.

```python
# 10년 단위 (1900, 1910, 1920, ...) 평균 기온의 평균 그려보기 - 온난화 확인
import csv
f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)
print(header)
# 년도 색인, 1900~1909 : 0, 1910 ~ 1919 : 1, ...
year = -1
# 평균 온도를 저장하는 13개의 리스트
avg_temp = []
min_temp = []
max_temp = []
for i in range(13) :
  avg_temp.append([])
  max_temp.append([])
  min_temp.append([])

for row in data :
# 빈 데이터가 아니라면 temp 리스트의 해당 요소(리스트)에 추가한다.
  # 일자를 '-' 단위로 분리한 첫 번째
  year = int(row[0].split('-')[0])
  # 일자를 10으로 나누고 190으로 나눈 나머지를 색인으로 사용
  idx = int(year/10) % 190
  if row[2] != '' : # 평균온도
    avg_temp[idx].append(float(row[2]))
  if row[4] != '' : # 최고온도
    max_temp[idx].append(float(row[4]))
  if row[3] != '' : # 최저온도
    min_temp[idx].append(float(row[3]))
```

```python
avgt = [0] * 13
maxt = [0] * 13
mint = [0] * 13
for i in range(13) :
  avgt[i] = sum(avg_temp[i]) / len(avg_temp[i])
  maxt[i] = sum(max_temp[i]) / len(max_temp[i])
  mint[i] = sum(min_temp[i]) / len(min_temp[i])
print(avgt)
print(maxt)
print(mint)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(12,6))
plt.plot(avgt, 'g', label='mean')
plt.plot(maxt, 'r--', label='max')
plt.plot(mint, 'b:', label='min')
#plt.ylim(-1,12)
plt.xticks(range(13), [1900 + i * 10 for i in range(13)], rotation=90)
plt.legend()
plt.show()
```

### 03. 기온 데이터를 히스토그램으로 표현하기, p82

```python
# p83, 기온데이터 히스토그램 그리기
import csv, matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
result = []

for row in data :
  if row[4] != '' :
    result.append(float(row[4]))

plt.hist(result, bins=100, color='r')
plt.show()
f.close()
```

```python
# p85, 1월과 8월의 기온데이터 히스토그램 그리기
import csv, matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
  if len(row) > 4 and row[4] != '' :
    month = row[0].split('-')[1]
    if month == '08' :
      aug.append(float(row[4]))
    if month == '01' :
      jan.append(float(row[4]))

plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
f.close()
```

```python
# p85, 1월과 8월의 기온데이터 히스토그램 그리기 - 다른 방법
import csv, matplotlib.pyplot as plt

f = open('seoul-utf8.csv', encoding='utf-8')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data :
  if len(row) == 5 and row[4] != '' :
    if '-08-' in row[0] :
      aug.append(float(row[4]))
    if '-01-' in row[0] :
      jan.append(float(row[4]))

plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
f.close()
```
